```python
TARGET_VALUE = 100

def get_all_expressions(exp, expressions, numbers, operations, idx):
    for op in operations:
        cur_exp = exp[:]
        cur_exp += str(numbers[idx])
        if idx < len(numbers) - 1:
            cur_exp += op
            get_all_expressions(cur_exp, expressions, numbers, operations, idx+1)
        elif eval(cur_exp) == TARGET_VALUE and cur_exp not in expressions:
            expressions.append(cur_exp)

def solve_hundred_problem(num_solutions = 10):
    numbers = [x for x in range(1, 10)]
    operations = ['*', '+', '-', '/', '']
    expressions = []
    get_all_expressions("", expressions, numbers, operations, 0)
    print('Num solutions: ', len(expressions))
    print(f'First {num_solutions} solutions: ')
    [print(expressions[idx]) for idx in range(num_solutions)]
    
solve_hundred_problem(num_solutions=25)
```

```python
boards = []
def place_queens(size, boards, board, x):
    for i in range(0, size):
        b = [row[:] for row in board]
        if b[i][x] == 1:
            continue
        place_queen(size, b, x, i)
        if x < size-1:
            place_queens(size, boards, b, x+1)
        else:
            boards.append(b)
            
def place_queen(size, board, x, y):
    for field in range(0, size):
        board[field][x] = 1
        board[y][field] = 1
    i = 1
    while x-i >= 0 and y-i >= 0:
        board[y - i][x - i] = 1
        i+=1
    i = 1
    while x+i < size and y+i < size:
        board[y + i][x + i] = 1
        i+=1
    i = 1
    while x-i >= 0 and y+i < size:
        board[y + i][x - i] = 1
        i+=1
    i = 1
    while x+i < size and y-i >= 0:
        board[y - i][x + i] = 1
        i+=1
    board[y][x] = 8

        
def print_board(size, board):
    print("  ", end='  ')
    x_names = [print(chr(x), end=': ') for x in range(65, 65+size)]
    print()
        
    for i in range(size):
        print(i+1, end=': ')
        print(board[i])
    print("")
    
def print_solution(size, board):
    print(f'Possible solution for {size} queen problem: ')
    x_names = [chr(x) for x in range(65, 65+size)]
    for x in range(size):
        for y in range(size): 
            if board[y][x] == 8:
                print(f'Queen at ({x_names[x]}|{y + 1})')
                

def solve_queens_problem(size):
    boards = []
    board = [[0 for y in range(size)] for x in range(size)]
    place_queens(size, boards, board, 0)
    print(f'Number of solutions for {size} queen problem: {len(boards)}')
    if len(boards) != 0:
        print_solution(size, boards[0])
        print_board(size, boards[0])

solve_queens_problem(3)
solve_queens_problem(4)
solve_queens_problem(8)
solve_queens_problem(10)
```
