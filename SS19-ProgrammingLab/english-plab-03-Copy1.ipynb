```python
from itertools import combinations_with_replacement
import itertools
```

```python

cds = [1,2,5,10,20,50] #variable name: chnge_denominations
vc = set() #var name: valid_change
for i in range(2,100):
    for r in range(i):
        for comb in combinations_with_replacement(cds,r):
            if sum(comb) == 100:
                vc.add(comb)
```

<u>Smaller problem, how many smaller coins go into 10</u>

```python
tally = 0
iteration = 0
def R(bd,scl): #func name = recursion, var name = break_down, var name = smaller_coin_list
    
    global iteration
    iteration +=1
    print("\n\niteration:{}".format(iteration))
    global tally
    if scl: 
        mohd = int(bd/scl[0]) #var name = multiples_of_highest_denomination
        for i in range(0,mohd+1):
            print("i:{}\nscl:{}\nbd:{}\nmohd:{}\nmohd*i:{}\nbd-(mohd*i):{}\ntally: {}\n".format(i,scl,bd,mohd,(mohd*i),bd-(mohd*i),tally))
            if (bd-(mohd*i)) == 0 : 
                tally+= 1
            if scl[0] != 1 and (bd-(mohd*i)) > 0: R(bd-(mohd*i),scl[1:])

    
R(10,[5,2,1])
print("tally: {}".format(tally))
```

```python
import numpy as np
from itertools import product
count = 0
y1 = np.arange(101)
y2,y3,y4,y5,y6 = y1[0:50],y1[0:20],y1[0:10],y1[0:5],y1[0:2]
a,b,c,d,e,f =y1*1,y2*2,y3*5,y4*10,y5*20,y6*50
x = list(itertools.product(a,b,c,d,e,f))
for each in x:
    if sum(each) == 100: count +=1
print(count)
```
