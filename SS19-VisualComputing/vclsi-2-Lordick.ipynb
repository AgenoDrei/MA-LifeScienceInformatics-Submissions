```python
import pandas as pd
df=pd.read_excel("Desktop/only_5_relevant_atributes.xlsx",index_col=0)
#df.drop("0, axis=1)
df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

g = sns.PairGrid(df,hue="class",vars=["thickness","uniCelS","uniCelShape","bareNuc","blaChroma"],palette="Set1",hue_kws={"marker": ["o", "s"]})

g = g.map_diag(plt.hist)
g = g.map_lower(sns.regplot,scatter=False,fit_reg=True)
g = g.map_upper(plt.scatter)
g = g.add_legend()

plt.show()
```
